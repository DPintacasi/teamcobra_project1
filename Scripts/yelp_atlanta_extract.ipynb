```python
city_ethnic_details = pd.DataFrame({"City":'Atlanta'},index = [0])
city_ethnic_details['Name'] = ''
city_ethnic_details['City'] = ''
city_ethnic_details['Latitude'] = ''
city_ethnic_details['Longitude'] = ''
city_ethnic_details['ZipCode'] = ''
city_ethnic_details['Ethnicity'] = ''
city_ethnic_details['Price Level'] = ''
city_ethnic_details['Rating'] = ''

city_ethnic_details
```

```python
#chinese calls for details
API_HOST = 'https://api.yelp.com/v3/businesses/search'
headers = {'Authorization': 'Bearer %s' % yelp_key}
params={'term':'restaurants','location':'Atlanta', 'categories':'chinese'}
    
atlanta_chinese = requests.get(API_HOST,headers=headers, params=params).json()

city_ethnic_details.loc[0, "Name"] = atlanta_chinese['businesses'][0]['name']
city_ethnic_details.loc[0, "City"] = atlanta_chinese['businesses'][0]["location"]["city"]
city_ethnic_details.loc[0, "Latitude"] = atlanta_chinese['businesses'][0]["coordinates"]["latitude"]
city_ethnic_details.loc[0, "Longitude"] = atlanta_chinese['businesses'][0]["coordinates"]["longitude"]
city_ethnic_details.loc[0, "ZipCode"] = atlanta_chinese['businesses'][0]["location"]["zip_code"]
#city_ethnic_details.loc[0, "Ethnicity"] = atlanta_chinese['businesses'][0]["categories"]["alias"]
#city_ethnic_details.loc[0, "Price Level"] = atlanta_chinese['businesses'][0]["price"]
city_ethnic_details.loc[0, "Rating"] = atlanta_chinese['businesses'][0]["rating"]

city_ethnic_details
```
